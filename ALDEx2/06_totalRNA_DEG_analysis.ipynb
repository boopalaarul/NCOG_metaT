**Joining mixo ORF annotations to DEGs**

```python
import pandas as pd
```

```python
#file too big. have to read it line by line.
annotation_columns = pd.read_csv("nov2024_totalRNA_RAPresults/NCOG-totalRNA-annotations.tsv", sep = "\t", index_col=0,
                                  nrows=1, header = None).values.tolist()[0]
```

```python
len(annotation_columns)
```

```python
upreg_DEGs = pd.read_csv("05_out_upreg_pval.tsv", sep = "\t", index_col = 0)
downreg_DEGs = pd.read_csv("05_out_downreg_pval.tsv", sep = "\t", index_col = 0)
```

```python
upreg_DEGs[list(annotation_columns)] = ""
downreg_DEGs[list(annotation_columns)] = ""
```

```python
#large file, read line by line
for line in open("nov2024_totalRNA_RAPresults/NCOG-totalRNA-annotations.tsv", "r"):
    terms = line.strip("\n").split("\t")
    assert(len(terms) == 19)
    contig_id = terms[0]
    annotations = terms[1:]
    if contig_id in upreg_DEGs.index:
        upreg_DEGs.loc[contig_id, annotation_columns] = annotations
    elif contig_id in downreg_DEGs.index:
        downreg_DEGs.loc[contig_id, annotation_columns] = annotations
```
